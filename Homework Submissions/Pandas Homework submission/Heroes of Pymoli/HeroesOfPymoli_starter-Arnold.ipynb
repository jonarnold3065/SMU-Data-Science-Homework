```python
# create the bins 
bins = [0, 9, 14, 19, 24, 29, 34, 39, 10000]


age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_groups
```

```python
# add age group to summary table 
player_data["age_groups"] = pd.cut(player_data["Age"], bins, labels=age_groups, include_lowest=True)
player_data
```

```python
player_data["age_groups"].value_counts()
```

```python
age_breakout  = pd.DataFrame(player_data.groupby("age_groups").size(), columns=["Count"])
age_breakout["Percentage"] = age_breakout["Count"] / len(player_data) * 100
#gender_data.round({'Percentage': 1})
age_breakout.round({'Percentage': 2}).style.format({"Percentage": "%{:20}"})
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
sales_data["AgeGroup"] = pd.cut(sales_data["Age"], bins, labels=age_groups, include_lowest=True)
sales_data.head()
```

```python
a_purch_count = sales_data.groupby("AgeGroup").size()
a_avg_purch_price = sales_data.groupby("AgeGroup").Price.mean()
a_price_sum = sales_data.groupby("AgeGroup").Price.sum()
```

```python
age_purch_summ = pd.DataFrame([a_purch_count, a_avg_purch_price, a_price_sum]).T
age_purch_summ.columns = ["Total Purchases", "Average Purchases", "Sum Purchases"]
age_purch_summ['Purchases by Player Avg'] = age_purch_summ["Sum Purchases"] / age_breakout["Count"]

age_purch_summ
```

```python
age_purchasing = pd.DataFrame([a_purch_count, a_avg_purch_price, a_price_sum]).T
age_purchasing.columns = ['Total Purchases', 'Avg. Purchase Price', 'Sum of Purchases']
age_purchasing['AVG Purchases by Player'] = age_purchasing['Total Purchases'] / age_breakout['Count']

age_purchasing.round({'Avg. Purchase Price': 2, 
                      "Purchases by Player Avg": 2}).style.format({"Average Purchase": "${:20}",
                    "Purchases by Player Avg": "${:20}"})
```
